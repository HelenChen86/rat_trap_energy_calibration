```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import os

def get_spec(file):
    with open(os.path.join('sample_spectrum', file), 'r') as f:
        lines=f.readlines()
        headerline_ind = 0
        for line in lines:
            if 'Energy_(eV)' in line:
                break
            headerline_ind += 1 
    f.close()
    data = pd.read_csv(os.path.join('sample_spectrum', file), sep="\t", skiprows=[x for x in range(headerline_ind)])
    return data
```
